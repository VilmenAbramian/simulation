Загрузим рузультаты симуляции из json-строки.
Внимание! В переменной path нужно указать путь,
в котором сохраняются результаты симуляции

```python
path = r'/Users/vilmen/Documents/PhD/rfidsim/pysim/models/monte_carlo/results/text'
files = os.listdir(path)
files = [os.path.join(path, file) for file in files]
files = [file for file in files if os.path.isfile(file)]
latest = max(files, key=os.path.getctime)
print(latest)
```

```python
f = open(latest, 'r')
lines = f.readlines()
f.close()

line = lines[1]
line = line[27:-1]
res_list = list(line.split(', '))
for i in range(len(res_list)):
    res_list[i] = float(res_list[i])
print(res_list)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), layout='constrained')
ax.plot(
    list(range(len(res_list))), res_list,
    linewidth=3, linestyle='-',
    marker='s', markevery=30,
    markersize=8,
)
```
